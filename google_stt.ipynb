```python
from google.cloud import speech_v1p1beta1 as speech    # speaker_diarization
from google.cloud.speech_v1p1beta1 import enums
from google.cloud.speech_v1p1beta1 import types
```

```python
client = speech.SpeechClient()
```

```python
# gcs_uri = "gs://voice_magic_audio/황금종려상자랑봉준호_3명인터뷰_YTN_16000.wav"    # 한글 파일명 인식 오류
gcs_uri = "gs://voice_magic_audio/bong_interview_YTN_16000.wav"
```

```python
audio = types.RecognitionAudio(uri=gcs_uri)
```

```python
config = types.RecognitionConfig(
            encoding=enums.RecognitionConfig.AudioEncoding.LINEAR16,
            sample_rate_hertz=16000,
            language_code='ko-KR',
            enable_speaker_diarization=True,
            diarization_speaker_count=3)
```

```python
response = client.long_running_recognize(config, audio)
```
